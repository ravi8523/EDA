```python
  order_keys=file_name['job_title'].value_counts().keys()
order_keys
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=file_name,x='job_title',order=order_keys)

plt.xlabel("job_title")
plt.ylabel("count")
plt.title("count plot")
```

```python
file_name['company_location'].value_counts()
name=file_name['company_location'].value_counts().keys()
values=file_name['company_location'].value_counts().to_list()
print(name)
print(values)
company_location_df=pd.DataFrame(zip(name,values),columns=['type','count'])
```
